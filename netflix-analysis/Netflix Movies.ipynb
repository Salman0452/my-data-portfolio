```python
sns.heatmap(df.isna(), ax=plt.axes())
plt.title("Missing Values", fontsize=12)
plt.xlabel("Columns", fontsize = 10) 
plt.ylabel("Missing Values", fontsize = 10)
plt.show()
```

```python
df_clean = df.copy()
df_clean
```

```python
df_clean['director'] = df_clean['director'].fillna("Unknown")
df_clean['cast'] = df_clean['cast'].fillna('Unknown')
df_clean['country'] = df_clean['country'].fillna('Unknown')
df_clean["date_added"] = pd.to_datetime(df_clean["date_added"], errors='coerce')
df_clean
```

```python
df_clean["type"].value_counts()
```

```python
top_countries = df_clean["country"].value_counts().head(10)
top_countries
```

```python
df_clean["listed_in"].value_counts().head(1)
```

```python
release_2010 = df_clean[df_clean["release_year"] >= 2010]
titles_per_year = release_2010['release_year'].value_counts()
titles_per_year
```

```python
avg_release_year = df_clean.groupby("type")["release_year"].mean()
avg_release_year
```

```python
df_clean[(df_clean["release_year"] > 2018) & (df_clean["country"] == "India")]
```

```python
df_clean[(df_clean["title"] == "TV Show") & (df_clean["listed_in"] == "Children & Family Movies")]
```

```python
love_titles = df[df['title'].str.contains('love', case=False)]
love_titles
```

```python
df_clean["year_added"] = df_clean["date_added"].dt.year
df_clean
```

```python
titles_added = df_clean["year_added"].value_counts()
titles_added
```

```python
avg_release_country = df_clean.groupby('country')['release_year'].mean()
avg_release_country.sort_values(ascending=False)
```
